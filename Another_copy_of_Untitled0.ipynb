```python
import pandas as pd

# Load dataset
data = pd.read_csv('Home Remedies.csv')

# Function to find matching remedy for given symptom
def find_remedy(symptom):
    matching_row = data[data['Health Issue'].str.contains(symptom, case=False)]
    if not matching_row.empty:
        return matching_row['Home Remedy'].iloc[0]
    else:
        return "No matching remedy found for the given symptom."

# Main function
def main():
    # User input
    symptom = input("What symptoms are you experiencing? ")

    # Find matching remedy
    remedy = find_remedy(symptom)

    # Output
    print("Remedy for {}:".format(symptom))
    print(remedy)

if __name__ == "__main__":
    main()
```

```python
import pandas as pd

# Assuming you have your dataset stored in a CSV file named hospitals.csv
# Load the dataset
df = pd.read_csv("HospitalsInIndia.csv")

def search_hospital(city_or_state):
    # Check if the input is a city name or state name
    if city_or_state in df['City'].unique():
        result = df[df['City'] == city_or_state]
    elif city_or_state in df['State'].unique():
        result = df[df['State'] == city_or_state]
    else:
        return "No hospitals found in the specified city or state."

    # Display the results
    for index, row in result.iterrows():
        print("Hospital Name:", row['Hospital'])
        print("Local Address:", row['LocalAddress'])
        print("Pincode:", row['Pincode'])
        print("-----------------------------")

# Example usage:
city_or_state = input("Enter the name of the city or state: ")
search_hospital(city_or_state)
```
